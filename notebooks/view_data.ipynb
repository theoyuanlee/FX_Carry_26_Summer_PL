```python
import pandas as pd
```

```python
ls  = [
    'em_fx_options_long',
    'em_fx_spot_forward_long',
    'em_interest_rates_long',
    'g10_fx_options_long',
    'g10_fx_spot_forward_long',
    'g10_interest_rates_long',
    'global_risk_long',
    'macro_market_proxies_long',
]

ls = [f'{i}.parquet' for i in ls]

tickers = []
```

```python
for i in ls:
    df = pd.read_parquet(f"../data/raw/{i}")
    for ticker in df['ticker'].unique():
        start = df.loc[df['ticker']==ticker]['date'].values[0]
        end = df.loc[df['ticker']==ticker]['date'].values[-1]
        len_df = len(df[df['ticker'] == ticker])
        tickers.append([ticker, start ,end, len_df])
        print(ticker, start ,end, len_df)

df = pd.DataFrame(tickers, columns = ['ticker','start','end','size'])
```

```python
df.to_csv("outputs/data_timerange.csv", index = None)
```
